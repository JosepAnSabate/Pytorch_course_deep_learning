# Reshaping, stacking, squezzing ans unsqueziing

# Reshaping
Reshapes a tensor to a defined shape

```python
import torch

x = torch.arange(1., 10.)
x, x.shape
```

```python
# add extra dim
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

# Stack
Combine multiple tensors

```python
x_stacked = torch.stack([x,x,x,x], dim= 0)
x_stacked, x_stacked.ndim, x_stacked.shape
```

```python
x_stacked = torch.stack([x,x,x,x], dim= 1)
x_stacked, x_stacked.ndim, x_stacked.shape
```

# Squeeze
removes all 1 dim

```python
x_reshaped, x_reshaped.shape, x_reshaped.ndim
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed, x_squeezed.shape, x_squeezed.ndim
```

# Unsqueeze

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed, x_unsqueezed.shape, x_unsqueezed.ndim
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
x_unsqueezed, x_unsqueezed.shape, x_unsqueezed.ndim
```
